```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
```

```python
url = "http://www.quanxue.cn/CT_DaoJia/LaoZi/LaoZi01.html"

req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
page_bytes = urlopen(req).read()
html = BeautifulSoup(page_bytes)
print(html.prettify()[2500:])
```

```python
my_p = html.find("p",{"class":"jingwen"})
print(my_p.getText())
```

```python
my_p = html.find("p",{"class":"comment"})
print(my_p.getText())
```

```python
my_p = html.find("div",{"class":"yinyong"})
print(my_p.getText())

```
